# <center> Iron Condor
* Four-Legged Option Strategy  
    1) Set a Short Strangle by Selling OTM Call and Put  
    2) Buy Further OTM Call (Against massive market rally)  
    3) Buy Further OTM Put (Against massive market decline)  
* Improved Short Strangle
    
# Market Expectation : Neutral

```python
# Market Expiry Price
MEP = np.arange(17500, 19700, 100)
#------------------------------------
# strike price (Short Position @ OTM CALL )
LST1 = 18650
Pre_Rec1 = 171
# Strike Rate (Short Position @ OTM PUT)
HST1 = 18350
Pre_Rec2 = 174
#------------------------------------
# Lower strike price (Long call @ OTM)
LST2 = 18750
Pre_Paid1 = 135
# Higher Strike Rate (Long put  @ OTM)
HST2 = 18250
Pre_Paid2 = 139
#------------------------------------
df = strategies.Iron_Condor(MEP,LST1,LST2,HST1,HST2,Pre_Rec1,Pre_Rec2, Pre_Paid1, Pre_Paid2)
df.head(len(MEP))
```

# <center> Butterfly Spread
    
## Combination of Bull Spread and Bear Spread:
    
## Buy 1 ITM Call

    Buy at lower Strike Price
    Closes at expiry
    Provides the Maximum Profit

## Sell 2 ATM Calls

    Sold at middle Strike Price
    Generates the revenue
    Expires worthless

## Buy 1 OTM Call

    Buy at higher Strike Price
    Expires worthless


```python
# Market Expiry Price
# underlying value 17895 Rs

MEP = np.arange(16000, 19000, 100)
#------------------------------------
# Lower strike price (Long Position @ ITM)
LST1 = 17200
Premium_Paid1 = 327
#------------------------------------
# Lower Strike  (Short Position @ ATM)
LST2 = 17550
Premium_Rec1 = 47
Premium_Rec2 = 47

#Higher Strike (Long Position @ OTM)
HST = 17900
Premium_Paid2 = 2.6

df = strategies.Butterfly_Spread(MEP,LST1,LST2,HST,Premium_Paid1, Premium_Paid2,Premium_Rec1,Premium_Rec2)
df.head(len(MEP))
```

# <center>  Iron Butterfly (Iron Fly)
    
## 1) One Bull Call -> Long Call OTM & Short Call ATM

## 2) One Bear Put  -> Long Put OTM & Short Put ATM

```python
MEP = np.arange(16000, 19000, 100)
# Bull Call 
#------------------------------------
#  (Long CALL @ OTM)
LST1 = 17900
Pre_Paid1 = 4.5

#------------------------------------
# (Short CALL @ ATM)
HST1 = 17600
Pre_Rec1 = 28.45

# Bear Put 
# (Long Put @ OTM )
LST2 = 17300
Pre_Paid2 = 47.1
#------------------------------------
# (Short Put @ ATM )
HST2 = 17600
Pre_Rec2 = 103.8
df = strategies.Iron_Fly(MEP,LST1,LST2,HST1,HST2,Pre_Rec1,Pre_Rec2, Pre_Paid1, Pre_Paid2)
df.head(len(MEP))
```
