# <center> Bull Put Spread</center>

   2) Sell 1 ITM Put Option -> Leg 2  
* Depends on premium one can choose either Bull Call or Bull Put Strategy  
* Spread Exceutes for Credit

```python
#------------------------------------
# Market Expiry Price
MEP = np.arange(17700, 19700, 100)
#------------------------------------
# Lower strike price (Long Put @ OTM)
LST = 18100
Premium_Paid = 39.5
#------------------------------------
# Higher Strike Rate (Short Put @ ITM)
HST = 18500
Premium_Received = 200
#------------------------------------
df = strategies.Bull_Put(MEP,LST,HST,Premium_Paid,Premium_Received)
df.head(len(MEP))
```
