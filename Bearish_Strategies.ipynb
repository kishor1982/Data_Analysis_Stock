# <center> Bear Put Spread
* Two Leg Spread using Long and Short Put  
    1) 1 Buy ITM Put Option   
    2) 1 Sell OTM Put Option    
# Market Expectation : Moderately Bearish (4-5% Correction)

```python
#------------------------------------
# Market Expiry Price
MEP = np.arange(16000, 18000, 100)
#------------------------------------
# (Long Put @ ITM )
LST = 17400
Premium_Paid = 187.50
#------------------------------------
# (Short Put @ OTM )
HST = 16900 
Premium_Received = 49.95
#------------------------------------
df = strategies.Bear_Put(MEP,LST,HST,Premium_Paid,Premium_Received)
df.head(len(MEP))
```
